```python
from cornac.datasets import citeulike
import sys
sys.path.append('../')
import recs
```

```python
data = citeulike.load_feedback()
data
```

```python
user, item, rate = [], [], []

for i in data:
    user.append(i[0])
    item.append(i[1])
    rate.append(i[2])



import pandas as pd
df = pd.DataFrame({
    "user": user,
    "item": item,
    "rate": rate,
})

df.drop_duplicates(inplace=True)
df.info()
```

```python
from recs.models.mf import MatrixFactorization
```

```python
MF = MatrixFactorization(df, 15)
```

```python
from tqdm import tqdm
for u in tqdm(set(user)):
    MF.rec_items(u)[MF.pivot_columns].tolist()
```
